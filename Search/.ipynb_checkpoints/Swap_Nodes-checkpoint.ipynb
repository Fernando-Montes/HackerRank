```python
# Swap Nodes
import sys

class node:
    def __init__(self, parent, left = None, right = None, depth = 0):
        self._parent = parent
        self._left = left
        self._right = right
        self._depth = depth
        
    def setChildren(self, left, right):
        self._left = left
        self._right = right
        
    def getParent(self):
        return self._parent
    
    def getChildren(self):
        return self._left, self._right
    
    def getDepth(self):
        return self._depth
    
class queue:
    def __init__(self):
        self._q = []
        
    def enqueue(self, value):
        self._q.append(value)
        
    def dequeue(self, last = False):
        if last:
            i = -1
        else:
            i = 0
        tmp = self._q[i]
        self._q.pop(i)
        return tmp
    
    def empty(self):
        if len(self._q) == 0:
            return True
        else:
            return False
    
def breadthFirstInput(tree, left, right): 
    '''
    Inputs data into a tree using a bread first approach
    Creates queues of nodes to add children
    '''
    q = queue()    
    q.enqueue(tree)   
    setC = 0
    while (q.empty() != True and setC == 0) :
        n = q.dequeue()
        if n.getParent() != -1:  # This is since there are some lines missing in input
            nleft, nright = n.getChildren()
            if nleft == None and nright == None:   # Found empty node!
                n.setChildren( node(left,  depth = n.getDepth() + 1), \
                               node(right, depth = n.getDepth() + 1) )
                setC = 1
            else:               # Not empty so add them to queue 
                q.enqueue(nleft)
                q.enqueue(nright)
    return tree

def depthFirstOutput(tree):
    '''
    Returns tree data using a depth first approach
    '''
    left, right = tree.getChildren()
    res = []
    if left != None:
        res.append( depthFirstOutput( left ) )
    if tree.getParent() != -1: 
        res.append( tree.getParent() ) 
    if right != None:
        res.append( depthFirstOutput( right ) )
    return res
            
def fillTree(arr):
    tree = node(1)
    for i in range(len(arr)):
        tree = breadthFirstInput(tree, arr[i][0], arr[i][1])
    return tree  
    
def breadthFirstSwap(tree, k): 
    q = queue()
    q.enqueue(tree) 
    while q.empty() != True:
        n = q.dequeue()
        if n.getParent() != -1:
            nleft, nright = n.getChildren()
            q.enqueue(nleft)
            q.enqueue(nright)
            if (n.getDepth()+1)%k == 0:
                n.setChildren(nright, nleft)
    return tree
    
def swapNodes(indexes, queries):
    tree = fillTree(indexes)    
    result = []
    for i in queries:
        tree = breadthFirstSwap(tree, i)
        #tmp = str(depthFirstOutput(tree)).strip()
        #res = [int(tmp[i]) for i in range(len(tmp)) if tmp[i].isdigit() ]
        tmp = depthFirstOutput(tree)
        tmp = str(tmp).split(', ')
        res = [tmp[i] for i in range(len(tmp)) if str(tmp[i]).isdigit() ]
        result.append( res )
    return result
    
sys.setrecursionlimit(10000)
n = int(input())
indexes = []
for _ in range(n):
    indexes.append(list(map(int, input().rstrip().split())))
queries_count = int(input())
queries = []
for _ in range(queries_count):
    queries_item = int(input())
    queries.append(queries_item)
result = swapNodes(indexes, queries)
print('\n'.join([' '.join(map(str, x)) for x in result]))
```
