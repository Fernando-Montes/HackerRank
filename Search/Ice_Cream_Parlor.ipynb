```python
def binarySearch(arr, num, left, right):
    #print (num, left, right)
    if left == right:
        if arr[left] == num:
            return left
        else:
            return False
    elif left + 1 == right:
        if arr[right] == num:
            return right
        elif arr[left] == num:
            return left
        else:
            return False
    else:
        m = int((left + right)/2)
        if arr[m] == num:
            return m
        elif arr[m] < num:
            left = m
            return binarySearch(arr, num, left, right)
        else:
            right = m
            return binarySearch(arr, num, left, right)
    

def whatFlavors(cost, money):
    d = {str(cost[i]): i for i in range(len(cost))}
    costSorted = sorted(cost)
    Found = 0
    i = 0
    while Found == 0 and i < len(costSorted)-1:
        tmp = binarySearch(costSorted, money-costSorted[i], i+1, len(costSorted)-1)
        if tmp != False:
            Found = 1
            if d[str(costSorted[i])] < d[str(costSorted[tmp])] :
                print(d[str(costSorted[i])]+1, d[str(costSorted[tmp])]+1)
            elif d[str(costSorted[i])] > d[str(costSorted[tmp])]:
                print(d[str(costSorted[tmp])]+1, d[str(costSorted[i])]+1)
            else: 
                tmp = [j for j in range(len(cost)) if cost[j] == costSorted[i]]
                print(tmp[0]+1, tmp[1]+1)
        i = i + 1

t = int(input())
for t_itr in range(t):
    money = int(input())
    n = int(input())
    cost = list(map(int, input().rstrip().split()))
    whatFlavors(cost, money)
```
