```python
# https://www.hackerrank.com/challenges/pacman-dfs/problem
# https://www.hackerrank.com/challenges/pacman-dfs/forum/comments/131659

def inputData():
    player = [int(i) for i in input().strip().split()]
    food = [int(i) for i in input().strip().split()]
    size = [int(i) for i in input().strip().split()]
    board = [[j for j in input().strip()] for i in range(size[0])]
    return player, food, size, board

class myStack(): #FILO: First In Last Out
    def __init__(self):
        self._s = []  
    def insert(self, pos):
        self._s.insert(0, pos)
    def pop(self):
        tmp = self._s[0]
        self._s.pop(0)
        return(tmp)
    def firstOut(self):
        return(self._s[0])
    def isEmpty(self):
        return( len(self._s) == 0 )

def insertStack(stack, pos, board, added, path):
    if board[pos[0]-1][pos[1]] != '%' and [pos[0]-1, pos[1]] not in added:  # insert Up
        stack.insert( [pos[0]-1, pos[1]] )
        added.append( [pos[0]-1, pos[1]] )
        if repr([pos[0]-1, pos[1]]) not in list(path.keys()):
            path[ repr([pos[0]-1, pos[1]]) ] = [pos[0], pos[1]]
    if board[pos[0]][pos[1]-1] != '%' and [pos[0], pos[1]-1] not in added:  # insert Left
        stack.insert( [pos[0], pos[1]-1] )
        added.append( [pos[0], pos[1]-1] )
        if repr([pos[0], pos[1]-1]) not in list(path.keys()):
            path[ repr([pos[0], pos[1]-1]) ] = [pos[0], pos[1]]
    if board[pos[0]][pos[1]+1] != '%' and [pos[0], pos[1]+1] not in added:  # insert Right
        stack.insert( [pos[0], pos[1]+1] )
        added.append( [pos[0], pos[1]+1] )
        if repr([pos[0], pos[1]+1]) not in list(path.keys()):
            path[ repr([pos[0], pos[1]+1]) ] = [pos[0], pos[1]]
    if board[pos[0]+1][pos[1]] != '%' and [pos[0]+1, pos[1]] not in added:  # insert Down
        stack.insert( [pos[0]+1, pos[1]] )
        added.append( [pos[0]+1, pos[1]] )
        if repr([pos[0]+1, pos[1]]) not in list(path.keys()):
            path[ repr([pos[0]+1, pos[1]]) ] = [pos[0], pos[1]]
    return stack, added, path
    
def explore(player, food, size, board):
    stack = myStack()
    added = []
    path = {}
    stack, added, path = insertStack(stack, player, board, added, path)
    done = False
    explored = [player]
    while stack.isEmpty() == False and done == False:
        pos = stack.pop()
        explored.append(pos)
        if pos == food:
            done = True
        else:
            stack, added, path = insertStack(stack, pos, board, added, path)
    return explored, added, path

#player, food, size, boardd = inputData()
board = boardd
explored, added, path = explore(player, food, size, board)

print(len(explored))
for pos in explored:
    print(pos[0], pos[1])

pathStack = myStack()
pathStack.insert( path[repr(food)] ) 
pathLength = 0
while pathStack.firstOut() != player:
    pathLength = pathLength + 1
    pathStack.insert( path[repr( pathStack.firstOut() )] )  
print(pathLength+1)
for i in range(pathLength+1):
    tmp = pathStack.pop()
    print( tmp[0], tmp[1] )
print( food[0], food[1])
```
