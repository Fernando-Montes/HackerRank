```python
# The captcha cracker
# https://www.hackerrank.com/challenges/the-captcha-cracker/problem

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Creation of dictionary with the all the characters and numbers
# Reduce all RGB to single 0 or 1 character
data = {}
for i in range(0,25) :
    fileIN = pd.read_table('~/Dropbox/Courses/Python/Algorithms/HackerRank/Data/sampleCaptchas/input/input' + str(i).zfill(2) + '.txt', header = None)
    fileOUT = pd.read_table('~/Dropbox/Courses/Python/Algorithms/HackerRank/Data/sampleCaptchas/output/output' + str(i).zfill(2) + '.txt', header = None)
    dim = [int(i) for i in fileIN[0][0].strip().split()]

    m = np.asarray([[0 for i in range(dim[1])] for j in range(dim[0])])
    for r in range(0, dim[0]):
        tmp1 = [i for i in fileIN[0][1 + r].strip().split()]
        for c in range(0, dim[1]) :
            tmp2 = np.average([int(i) for i in tmp1[c].split(',')])
            if tmp2 > 150:
                m[r, c] = 0
            else:
                m[r, c] = 1
    
    charOut = [char for char in fileOUT[0][0]]
    for j in range(0,5):
        if charOut[j] not in data.keys():
            data[charOut[j]] = m[10:22, (4+9*j):(13+9*j)]
```
