```python
# Stock predictions

def trainInput():
    inputfile = open("Data/bigger_data_set.txt", "r")
    stock_prices = {}   
    for line in inputfile:
        tmp = list(map(str, line.split()))
        stock = tmp[0]
        prices = list(map(float, tmp[1:]))    
        stock_prices[stock] = prices
    return stock_prices

def plotPrice(prices, prices_pred = None):
    import matplotlib.pyplot as plt
    %matplotlib inline 
    plt.plot(prices)
    if len(prices_pred) != 0 :
        plt.plot(prices_pred)
    plt.xlabel('days')
    plt.ylabel('prices')
    plt.show()

# ---------------------------------------------
import math
import numpy as np
from sklearn.linear_model import LogisticRegression
num_features = 5
    
def readInput():
    m, k, d = map(float, input().split())
    stock_owned = {}
    try:
        stock_prices = np.load('stock_prices.npy').item()
    except:
        stock_prices = {}
    for i in range(int(k)):
        tmp = list(map(str, input().split()))
        stock = tmp[0]
        owned = int(tmp[1])
        prices = list(map(float, tmp[2:]))
        stock_owned[stock] = owned
        if stock in stock_prices.keys():
            stock_prices[stock].append(prices[-1]) 
        else:
            stock_prices[stock] = prices
    np.save('stock_prices.npy', stock_prices) 
    return m, k , d, stock_owned, stock_prices

def prepare(prices, num_features):
    x = []
    y = []
    for i in range(num_features, len(prices)):
        x.append(prices[(i-num_features):i])
        if prices[i] >= prices[i-1]:
            y.append(1)
        else:
            y.append(0)
    return x, y

def prediction(prices, num_features, model, pastFlag = False):
    x, y = prepare(prices, num_features)
    if pastFlag:
        x_pred = x
    else:
        x_pred = [prices[-5:]]
    move = model.fit(x,y).predict(x_pred)
    return move, y
 
def moves(stock_prices, num_features):
    direction = {}
    for stock in stock_prices.keys():
        if len(stock_prices[stock]) > 2*num_features+1:
            prices = stock_prices[stock]
            model = LogisticRegression(solver="newton-cg", random_state=42, penalty = 'l2') 
            move, y = prediction(prices, num_features, model, pastFlag = False)
            direction[stock] = move
    return direction

def decision(direction, m, stock_owned, stock_prices):
    sell = {}
    buy = {}
    all_stocks = list(direction.keys())
    for stock in all_stocks:
        if direction[stock] == 1 and stock_owned[stock] != 0: # If want to buy but already owned remove
            direction.pop(stock)
        elif direction[stock] == 0 and stock_owned[stock] == 0: # If want to sell but do not own remove
            direction.pop(stock)
    all_stocks = list(direction.keys())
    for stock in all_stocks:  # Making list of sells
        if direction[stock] == 0:
            sell[stock] = stock_owned[stock]
            direction.pop(stock)    
    all_stocks = list(direction.keys())
    if len(direction) != 0:
        m_stock = m/len(direction)  # Money willing to invest per stock
    for stock in all_stocks:  # Making list of buys
        if m_stock/stock_prices[stock][-1] > 1:
            buy[stock] = math.floor(m_stock/stock_prices[stock][-1])
            m = m - buy[stock]*stock_prices[stock][-1]
    for stock in all_stocks:  # Use the rest of the money for buying
        if m/stock_prices[stock][-1] > 1:
            buy[stock] = math.floor(m/stock_prices[stock][-1])
            m = m - buy[stock]*stock_prices[stock][-1]
    return sell, buy

m, k ,d, stock_owned, stock_prices = readInput()
#stock_prices = trainInput()
direction = moves(stock_prices, num_features)
sell, buy = decision(direction, m, stock_owned, stock_prices)

print(len(sell)+len(buy))
for stock in sell.keys():
    print(str(stock) + " SELL " + str(sell[stock]))
for stock in buy.keys():
    print(str(stock) + " BUY " + str(buy[stock])) 
```
