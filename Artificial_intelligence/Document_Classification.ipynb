```python
# Document Classification
# Alternative 1 (slow and not complete):

import pandas as pd
import numpy as np

def readInput():
    inputfile = open("Data/trainingdata.txt", "r")
    num_tests  = int(inputfile.readline())
    x = []
    y = []
    for i in range(num_tests):
        tmp = list(map(str, inputfile.readline().split()))
        y.append(int(tmp[0]))
        x.append(tmp[1:])
    return num_tests, x, y

def createDict(num_tests, x):
    word2num = {}
    num2word = {}
    num_words = 0
    countWords = []
    for i in range(num_tests):
        for j in range(len(x[i])):
            if x[i][j] not in word2num:
                word2num[x[i][j]] = num_words
                num2word[num_words] = x[i][j]
                countWords.append(1) 
                num_words = num_words + 1
            else:
                countWords[word2num[x[i][j]]] = countWords[word2num[x[i][j]]] + 1
    return word2num, num2word, countWords     
                
def createDataFrame(num_tests, vocab, x, word2num, num2word):
    columns = [num2word[k] for k in vocab]
    columns.insert(0, 'not_in_vocab')
    df = pd.DataFrame(columns = columns)
    df = df.fillna(0)
    total_words = len(vocab) + 1
    for i in range(num_tests):
    #for i in range(2):
        df.loc[i] = list(np.zeros(total_words, dtype = int))
        notVocab = 0
        for j in range(len(x[i])):
            if x[i][j] in columns:
                df.loc[i, x[i][j]] = df.loc[i, x[i][j]] + 1
            else:
                df.loc[i, 'not_in_vocab'] = df.loc[i, 'not_in_vocab'] + 1
    return df

num_vocab = 1000
num_tests, x, y = readInput()
word2num, num2word, countWords = createDict(num_tests, x)
vocab = np.argsort(countWords)[::-1]
vocab = vocab[:num_vocab]
df = createDataFrame(num_tests, vocab, x, word2num, num2word)
```

```python
# Alternative 2 (fast):
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import VotingClassifier
import numpy as np
import warnings
warnings.filterwarnings("ignore")

def readTrainInput():
    inputfile = open("Data/trainingdata.txt", "r")
    num_tests  = int(inputfile.readline())
    x = []
    y = []
    for i in range(num_tests):
        tmp = inputfile.readline()
        y.append(int(tmp[0]))
        x.append("".join(tmp[1:]))                 
    return num_tests, x, y

def readTestInput():
    num_tests  = int(input())
    x = []
    for i in range(num_tests):
        x.append(input())                 
    return num_tests, x

def training(num_vocab):
    num_tests, x, y = readTrainInput()
    vectorModel = TfidfVectorizer(max_features = num_vocab, ngram_range=(1,5))
    #vectorModel = CountVectorizer(max_features = num_vocab)
    vectorModel.fit(x)
    x = vectorModel.transform(x).toarray()
    log = make_pipeline( StandardScaler(), LogisticRegression(solver="newton-cg", random_state=42, multi_class='multinomial', penalty = 'l2', class_weight='balanced') )
    log.fit(x,y)
    nbM = make_pipeline( MultinomialNB(alpha=1.0, class_prior=None, fit_prior=True) )
    nbM.fit(x,y)
    nbG = make_pipeline( GaussianNB(var_smoothing=1e-06) )
    nbG.fit(x,y)
    sgd = make_pipeline( StandardScaler(), SGDClassifier(loss='hinge', penalty='l2', alpha=1e-3, max_iter=5000, tol = 1e-3, random_state=42) )
    sgd.fit(x,y)
    eclf = VotingClassifier(estimators=[('log', log), ('nbM', nbM), ('nbG', nbG), ('sgd', sgd)], voting='hard')
    eclf.fit(x,y)
    return vectorModel, eclf, log, nbM, nbG, sgd

num_vocab = 300
vectorModel, eclf, log, nbM, nbG, sgd = training(num_vocab)

num_tests, x, y = readTrainInput()
#num_tests, x = readTestInput()
x = vectorModel.transform(x).toarray()
res = eclf.predict(x)
#for i in res:
#    print(i)
#np.mean(res == y)
```
