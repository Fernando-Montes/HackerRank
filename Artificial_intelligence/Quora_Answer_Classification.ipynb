```python
# Quora Answer Classification
#https://www.hackerrank.com/challenges/quora-answer-classifier/problem

from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import VotingClassifier

def readTrainInput():
    n, m = map(int, input().split())
    x = []
    y = []
    q = []
    for i in range(n):
        tmp = list(map(str, input().split()))
        q.append(tmp[0])
        y.append(int(tmp[1]))
        f = []
        for j in range(m):
            f.append(float(tmp[2+j].split(':')[1]))
        x.append(f)                 
    return m, x, y, q

def readTestInput(m):
    n = int(input())
    x = []
    q = []
    for i in range(n):
        tmp = list(map(str, input().split()))
        q.append(tmp[0])
        f = []
        for j in range(m):
            f.append(float(tmp[1+j].split(':')[1]))
        x.append(f)                 
    return x, q

def training(x, y):
    log = make_pipeline( StandardScaler(), LogisticRegression(solver="newton-cg", random_state=42, multi_class='multinomial', penalty = 'l2', class_weight='balanced') )
    log.fit(x,y)
    clf = RandomForestClassifier(n_estimators=50, min_samples_leaf=5, random_state=0)
    clf.fit(x,y)
    sgd = make_pipeline( StandardScaler(), SGDClassifier(loss='hinge', penalty='l2', alpha=1e-3, max_iter=5000, tol = 1e-3, random_state=42) )
    sgd.fit(x,y)
    eclf = VotingClassifier(estimators=[('log', log), ('clf', clf), ('sgd', sgd)], voting='hard')
    eclf.fit(x,y)
    return eclf, log, clf, sgd

m, x_train, y_train, q_train = readTrainInput()
x_test, q_test = readTestInput(m)
eclf, log, clf, sgd = training(x_train, y_train)
res = eclf.predict(x_test)
for i in range(len(q_test)):
    if res[i] == 1:
        print(q_test[i], '+1')
    else:
        print(q_test[i], '-1')
```

```python
# Quora Answer Classification --- Downloaded set

from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
import numpy as np
import warnings
warnings.filterwarnings("ignore")

def readInput():
    inputfile = open("../Data/input00_QUORA.txt", "r")
    n, m = map(int, inputfile.readline().split())
    x_train = []
    y_train = []
    q_train = []
    for i in range(n):
        tmp = list(map(str, inputfile.readline().split()))
        q_train.append(tmp[0])
        y_train.append(int(tmp[1]))
        f = []
        for j in range(m):
            f.append(float(tmp[2+j].split(':')[1]))
        x_train.append(f)     
    n = int(inputfile.readline())
    x_test = []
    q_test = []
    for i in range(n):
        tmp = list(map(str, inputfile.readline().split()))
        q_test.append(tmp[0])
        f = []
        for j in range(m):
            f.append(float(tmp[1+j].split(':')[1]))
        x_test.append(f) 
    return m, x_train, y_train, q_train, x_test, q_test

def readOutput():
    inputfile = open("../Data/output00_QUORA.txt", "r")
    y_test = []
    q_test = []
    for line in inputfile:
        tmp = list(map(str, line.split()))
        q_test.append(tmp[0])
        y_test.append(int(tmp[1]))
    return y_test, q_test

def training(x, y):
    log = make_pipeline( StandardScaler(), LogisticRegression(solver="newton-cg", random_state=42, multi_class='multinomial', penalty = 'l2', class_weight='balanced') )
    log.fit(x,y)
    nbM = make_pipeline( MultinomialNB(alpha=1.0, class_prior=None, fit_prior=True) )
    nbM.fit(x,y)
    nbG = make_pipeline( GaussianNB(var_smoothing=1e-06) )
    nbG.fit(x,y)
    clf = RandomForestClassifier(n_estimators=50, min_samples_split=70, min_samples_leaf=50, random_state=0)
    clf.fit(x,y)
    sgd = make_pipeline( StandardScaler(), SGDClassifier(loss='hinge', penalty='l2', alpha=1e-3, max_iter=5000, tol = 1e-3, random_state=42) )
    sgd.fit(x,y)
    #eclf = VotingClassifier(estimators=[('log', log), ('nbM', nbM), ('nbG', nbG), ('sgd', sgd)], voting='hard')
    eclf = VotingClassifier(estimators=[('log', log), ('clf', clf), ('sgd', sgd)], voting='hard')
    eclf.fit(x,y)
    return eclf, log, nbM, nbG, clf, sgd

m, x_train, y_train, q_train, x_test, q_test = readInput()
eclf, log, nbM, nbG, clf, sgd = training(x_train, y_train)
y_train_pred = eclf.predict(x_train)
acc_train_pred = y_train == y_train_pred 
print("Train acc.: ", sum(acc_train_pred)/len(acc_train_pred))

y_test_pred = eclf.predict(x_test)
y_test, q_test2 = readOutput()
if q_test2 == q_test:
    acc_test_pred = y_test == y_test_pred 
print("Test acc.: ", sum(acc_test_pred)/len(acc_test_pred))

#for i in range(len(q_test)):
#    if res[i] == 1:
#        print(q_test[i], '+1')
#    else:
#        print(q_test[i], '-1')
```
